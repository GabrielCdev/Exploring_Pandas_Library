### 01. Importar biblioteca

### 02. Criar os dataframes data1, data2 e data3 baseados nos dados abaixo

```
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

## Séries temporais

### 01. Importar as bibliotecas necessárias

```
import pandas as pd
import numpy as np

# visualization
import matplotlib.pyplot as plt

%matplotlib inline
```

### 02. Importar dados do link https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv

```
df_apple_stock = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv')

df_apple_stock
```

### 03. Checar as informações das colunas

```
df_apple_stock.info()
```

### 04. Transformar a coluna 'Date' para o formato datetime.

```
df_apple_stock.Date = pd.to_datetime(df_apple_stock.Date, format='%Y-%m-%d')

df_apple_stock
```

### 05. Verificar se existe alguma data repetida no dataframe

```
df_apple_stock[df_apple_stock.duplicated(subset=['Date'])]
```

### 06.  Definir a coluna 'Date' como índice

```
df_apple_stock.set_index('Date')
```

### 07.  Ops... as observações estão ordenadas da data mais recente para a mais antiga. É importante que a série temporal comece da observação mais antiga para a mais recente, uma vez que o tempo passa em sentido crescente.

```
df_apple_stock = df_apple_stock.set_index('Date').sort_values(by='Date')
df_apple_stock
```

### 08. Vamos criar um novo dataframe com apenas a observação do último dia útil de cada mês. Dica: verifique a função resample para séries temporais na biblioteca Pandas.

```
new_df = df_apple_stock.resample('BM').sum()

new_df
```

### 09. Qual a diferença em dias da data mais antiga para a mais recente?

```
(df_apple_stock.index[-1] - df_apple_stock.index[0]).days
```

### 10.  Quantos meses nós temos na base?

```
df_month = len(df_apple_stock.resample('M').sum().index)

df_month
```

### 11. Exibir o gráfico da coluna 'Adj Close' do dataframe apple.


```
plt.bar(range(len(df_apple_stock['Adj Close'])), df_apple_stock['Adj Close'])
```

### 12. Exibir o gráfico da coluna 'Adj Close' do dataframe apple_month.

```
plt.bar(range(len(df_apple_stock.resample('M').sum().index)), df_apple_stock.resample('M').sum()["Adj Close"])
```
